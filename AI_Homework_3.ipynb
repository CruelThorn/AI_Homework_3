Ще използвам библиотеката sklearn от Python за съдаване на оценяващата функция.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import mglearn
from sklearn.neighbors import KNeighborsClassifier
from IPython.display import display
%matplotlib notebook
```

Зареждаме данните

```python
dataset = pd.read_excel('data/LearningDataSet_Dom3.xlsx')
dataset
```

```python
dataset.columns=["Value", "Characteristic", "Success"]
```

Слагаме на видовете характеристики цифрови еквиваленти

```python
dataset.Characteristic.replace(['Weak', 'Average ', 'Strong', 'Very Strong '], [1, 2, 3, 4], inplace=True)
```

```python
dataset
```

```python
X = dataset.drop('Success', axis=1)
y = dataset['Success']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, stratify=y)
```

Създаваме алгоритъма, като използваме част от данните за тестови

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

Тъй като не постигаме 100% точност на алгоритъма ще увеличим броя на съседите по които определяме принадлежността

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```
